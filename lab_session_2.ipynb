```python
import pandapower as pp
import pandapower.plotting as plot
import pandapower.networks as pn
import matplotlib.pyplot as plt

#create empty net
net = pp.create_empty_network(f_hz=50.0, sn_mva=100)
```

```python
#create buses
bus0 = pp.create_bus(net, vn_kv=22, name="Bus 0", type="b")
bus1 = pp.create_bus(net, vn_kv=22, name="Bus 1", type="b")
bus2 = pp.create_bus(net, vn_kv=11, name="Bus 2", type="b")
bus3 = pp.create_bus(net, vn_kv=220, name="Bus 3", type="b")
bus4 = pp.create_bus(net, vn_kv=220, name="Bus 4", type="b")
bus5 = pp.create_bus(net, vn_kv=110, name="Bus 5", type="b")
bus6 = pp.create_bus(net, vn_kv=110, name="Bus 6", type="b")

#create bus elements. The n types are those adjacents to a transformer
```

```python
pp.create_impedance(net, from_bus=bus0,to_bus=bus1,r_pu=0,x_pu=0.2, rft_pu=0,xft_pu=0.2,sn_mva=100, name="gen")
pp.create_impedance(net, from_bus=bus1,to_bus=bus3,r_pu=0,x_pu=0.2, rft_pu=0,xft_pu=0.2,sn_mva=100, name="trafo 1")
pp.create_impedance(net, from_bus=bus3,to_bus=bus4,r_pu=0,x_pu=0.1, rft_pu=0,xft_pu=0.1,sn_mva=100, name="lines 1")
pp.create_impedance(net, from_bus=bus4,to_bus=bus2,r_pu=0,x_pu=0.15, rft_pu=0,xft_pu=0.15,sn_mva=100, name="trafo 2")
pp.create_impedance(net, from_bus=bus1,to_bus=bus5,r_pu=0,x_pu=0.16, rft_pu=0,xft_pu=0.16,sn_mva=100, name="trafo 3")
pp.create_impedance(net, from_bus=bus5,to_bus=bus6,r_pu=0,x_pu=0.5, rft_pu=0,xft_pu=0.5,sn_mva=100, name="lines 1")
pp.create_impedance(net, from_bus=bus6,to_bus=bus2,r_pu=0,x_pu=0.2, rft_pu=0,xft_pu=0.2,sn_mva=100, name="trafo 4")
```
